```
#한문학과 2013310866 이성현

#고려사항
#1. 무조건 농부가 배를 타야한다.
#2. 각각 왼쪽, 또는 오른쪽에 있어야 하고 캐릭터는 4개이기 때문에, 16가지 경우의 수가 존재한다.(실제는 더 적음)
# ffff ffft fftf ftff tfff fftt ftft tfft tftf fttf ttff tttf ttft tftt fttt tttt 중에서
#염소 배추 같이 있으면 안되고, 늑대 염소 같이 있으면 안되기 때문에 실제 가능한 경우의 수는 16개보다 적은 것
# 불가능 한 것: tfff fttt fftt fttf ttff tfft

#다음의 map은 key: 현상황, value: 다음으로 실행할 수 있는 경우의 수
false = False
true = True

map = {(false, false, false, false): [(true, false, true, false)],  # 모두 오른쪽에 있는경우
       (false, false, false, true): [(true, true, false, true), (true, false, true, true)],  # 양배추만 왼쪽인 경우
       (false, false, true, false): [(true, true, true, false), (true, false, true, true), (true, false, true, false)],
       # 염소만 왼쪽인 경우
       (false, true, false, false): [(true, true, true, false), (true, true, false, true)],  # 늑대만 왼쪽인 경우
       (false, true, false, true): [(true, true, true, true)],  # 늑대,양배추가 오른쪽에 있는 경우 - 여기 오면 이제 goal!!!!
       (true, false, true, false): [(false, false, true, false)],  # 농부와 염소가 왼쪽에 있는 경우, 염소데리고 다시 오른쪽 오는 건 의미가 없음(loop)
       (true, true, true, false): [(false, false, true, false), (false, true, false, false)],  # 농부와 늑대와 염소가 왼쪽에 있는 경우

       (true, true, false, true): [(false, true, false, true), (false, false, false, true),
                                   (false, true, false, false)],  # 농부와 늑대와 양배추가 왼쪽에 있는 경우
       (true, false, true, true): [(false, false, false, true), (false, false, true, false)],  # 늑대만 왼쪽에 있는 경우
       (true, true, true, true): [(false, true, false, true)]  # 는 goalState이기 때문에 만들 필요가 없다.
       }



class Node:
    def __init__(self, c, p, d):
        self.current = c
        self.parent = p
        self.depth = d

    def solutionPath(self):  # 결과적으로 True가 되는 노드를 return 받을 테니, 그때 노드의 현재위치, 엄마노드,,,등등을 호출할것이다.
        path = self.current
        if self.depth == 0:
            return path
        else:
            print(list(path))
            return list(self.parent.solutionPath())


    def __str__(self):
        return 'Node ' + str(self.current) + ', depth = ' + str(self.depth)


class Problem:
    def __init__(self, i, g):
        self.initState = tuple(i) #(false, false, false, false)
        self.goalState = tuple(g) #(true, true, true, true)
        self.map = map #<class 'dict'>

    def isGoal(self, current):
        if current == self.goalState:
            return True
        else:
            return False

    def sucFN(self, current):
        return self.map.get(current, [])


class Stack:
    def __init__(self):
        self.list = []

    #Push
    def myPush(self, node):
        self.list.append(node)

    #Pop
    def myPop(self):
        return self.list.pop()

    #IsEmpty
    def isEmpty(self):
        if len(self.list) == 0:
            return True
        else:
            return False


class DFS:
    def __init__(self, p):
        self.numberGeneratedNodes = 0  # 새로 생긴 node의 수
        self.prob = p
        self.frontier = Stack()
        self.visited = []

    def expand(self, parent):
        children = []

        for i in self.prob.sucFN(parent.current):
            s = Node(i, parent, parent.depth + 1)
            if s not in self.visited:
                children.append(s)
                self.visited.append(s)

        self.numberGeneratedNodes += len(children)
        return children

    def solution(self):
        # 1.root node with initial state
        root = Node(self.prob.initState, None, 0)

        # 2.add the root node to frontier Stack
        self.frontier.myPush(root)

        # while stack is not empty
        while not self.frontier.isEmpty():

            parent = self.frontier.myPop()
            self.visited.append(parent.current)
            if self.prob.isGoal(parent.current):
                return parent
            # expand and add them to the stack
            expandedNodes = self.expand(parent)
            for i in expandedNodes:
                if i.current in self.visited:
                    pass
                else:
                    self.frontier.myPush(i)
        return False
```
